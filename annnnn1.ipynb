```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def hyperbolicTan(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
def softmax(x):
    scores=np.exp(x)
    probabilities = scores / np.sum(scores, axis = 1 , keepdims= True)
    return probabilities
```

```python
x = np.linspace(-5,5,100)
```

```python
sig_op = sigmoid(x)
hyper_op = hyperbolicTan(x)
relu_op = relu(x)
soft_ip = np.array([[2,1,0.5] , [1,3,2] , [0.5,1,2]])
soft_op = softmax(soft_ip)
```

```python
plt.figure(figsize=(10,6))
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.plot(x , sig_op , label="Sigmoid")
plt.title("Sigmoid Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.plot(x , hyper_op , label="HyperbolicTan" , color= "orange")
plt.title("HyperbolicTangent Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.plot(x,relu_op,label="relu")
plt.title("ReLu Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.plot(soft_ip , soft_op.T ,marker="o")
plt.title("Softmax Activation Function")
plt.xlabel("Scores")
plt.ylabel("Probabilities")
plt.legend(["Class 1 , Class 2, Class 3"])
plt.tight_layout()
plt.show()
```
